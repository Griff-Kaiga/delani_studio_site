```
# importing the pandas library
import pandas as pd
# importing the numpy library
import numpy as np
```

```
cells_geo = 'cells_geo - cells_geo.csv'
df5 = pd.read_csv(cells_geo, header=0)
df5= df5.rename(columns={'SITE_CODE':'SITE_ID'})
df5
```

```
# drop rows where cells in that row is NA
df5_cleaned = df5.dropna(how='all')
df5_cleaned
```

```
# drop columns if they contain missing values
df5.dropna(axis=1, how='all') 
```

Data loading for the three days datasets

first day dataset

```
# creating an excel file to the url
url = 'http://bit.ly/Telcom_dataset1'

# then loading the first sheet of excel file into our data frame
df = pd.read_csv(url, header=0)
# renaming the column name
df = df.rename(columns = {'PRODUTC': 'PRODUCT'})
df = df.rename(columns = {'DATETIME' :'DATE_TIME'})
df

```

Second Day Dataset

```
# creating an excel file to the url
url_2 = 'http://bit.ly/Telcom_dataset2'

# then loading the first sheet of excel file into our data frame
df2 = pd.read_csv(url_2, header=0)
# renaming the column names 
df2 = df2.rename(columns = {'DW_A_NUMBER': 'DW_A_NUMBER_INT'})
df2
df10 = df2.rename(columns = {'DW_B_NUMBER': 'DW_B_NUMBER_INT'})
df10
```

third day dataset

```
# creating an excel file to the url
url_3 = 'http://bit.ly/Telcom_dataset3'

# then loading the first sheet of excel file into our data frame
df3 = pd.read_csv(url_3, header=0)
df3 = df3.rename(columns = {'SIET_ID': 'SITE_ID'})
df11 = df3.rename(columns = {'CELLID': 'CELL_ID'})
df11
```

Merging the three days datasets

```
# merging the three data frames 
df_row = pd.concat([df, df10, df11])
df_row
# printing out the first 50 rows
df_row.head(50)
```

Data cleaning for the three days datasets

```
# Dropping rows where all cells in that row is NA
df_row_cleaned = df_row.dropna(how='all')
df_row_cleaned
```

```
# dropping columns if they contain missing values
df_row.dropna(axis=1, how='all')
```

```
# dropping nrows if they contain less than 5 observations
df_row.dropna(thresh=5)
```

```
# dropping columns that we dont need
df_row1 = df_row.drop('CELL_ON_SITE', axis=1)
df_row2 = df_row1.drop('COUNTRY_A', axis=1)
df_row3 = df_row2.drop('COUNTRY_B', axis=1)
df_row3
```

```
# merging the cells geo and the df_row tables
df20 = pd.merge(df_row3, df5, on ='SITE_ID')
df20
```
